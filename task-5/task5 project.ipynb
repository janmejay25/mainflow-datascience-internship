```python
#1.how many people have heart disease and how many people doesn`t have heart disease?

#getting values
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind = 'bar',color=["orchid","salmon"])
plt.title("heart disease values")
plt.xlabel("1 =heart disease; 0=no heart disease")
plt.ylabel("amount");
```

```python
#plot pie chart 
df.target.value_counts().plot(kind = 'pie',figsize=(8,6))
plt.legend(["disease","no disease"]);
```

```python
#2. people of which sex has most heart disease?

#0= female , 1=male
#target columns 0= no disease 1= disease

df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind = 'pie',figsize= (8,6))
plt.title('male female ratio')
plt.legend(['male','female']);
```

```python
pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x='target',data = df,hue = 'sex')
plt.title("heart disease frequency for sex")
plt.xlabel("0 = no heart disease, 1 = heart disease");
```

```python
#number of male is more than double in our dataset than female
```

```python
#3. people of which sex has which type of chest pain most?
df.cp.value_counts()
```

```python
#bar chart 
df.cp.value_counts().plot(kind = 'bar', color = ['salmon','lightskyblue','springgreen','khaki'])
plt.title('chest pain type VS count');
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind = 'bar', color =['coral','lightskyblue','plum','khaki'])
plt.title('type of chest pain for sex')
plt.xlabel('0 = female, 1 = male');
```

```python
#most of the male has type0 chest pain and least of male has type4 pain

# in the case of female type 0 and type 2 is almost same
```

```python
#4. people with which chest painare most pron to have heart disease?

pd.crosstab(df.cp,df.target)
```

```python
sns.countplot(x= 'cp',data = df,hue = 'target');
```

```python
#most of people who has type 0 chest pain has less chance of heart disease.

#now we`ll see the opposite for other types.

#create a distribution plot with normal diss. curve
sns.displot(x = 'age',data = df, bins = 30,kde = True);
```

```python
#58-59 years old are most in dataset.

#plot chart for maximum heart rate
sns.displot(x = 'thalach',data= df, bins= 30,kde = True, color = 'chocolate');
```

```python
#5 What is the average age of people with heart disease versus those without?

df.groupby('target')['age'].mean()
```

```python
plt.figure(figsize=(8, 6))
df.groupby('target')['age'].mean().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Average Age of People with and without Heart Disease')
plt.xlabel('Heart Disease Presence (0 = No, 1 = Yes)')
plt.ylabel('Average Age')
plt.xticks(rotation=0)
plt.show()
```

```python
#6 Which 'thal' (thalassemia) type is most associated with heart disease presence?

pd.crosstab(df.thal,df.target)
```

```python
pd.crosstab(df.thal,df.target).plot(kind = 'bar', color =['coral','lightskyblue','plum','khaki'])
plt.title('type of thal with heart disease')
plt.xlabel('0 = No heart disease, 1 = heart disease');
```

```python
#type 2 has most heart disease comapre to other types of thal
```

```python
#7. Do people with 'exang' (exercise-induced angina) have a higher likelihood of heart disease?

pd.crosstab(df.exang,df.target)
```

```python
sns.countplot(x= 'exang',data = df,hue = 'target');
```

```python
pd.crosstab(df.exang,df.target).plot(kind = 'bar',figsize= (8,6))
plt.title('possibility of heart disease with exang')
plt.legend(['yes','No']);
```

```python
#For people without exang , a larger portion has heart disease 
#For people with exang , heart disease is less common.
```
