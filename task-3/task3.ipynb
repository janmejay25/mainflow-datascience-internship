```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data = pd.read_csv('C:\\Users\\janmejay\\Downloads\\householdtask3.csv')
```

```python
data.head(10)
```

```python
#scatter plot with year against own
plt.scatter(data['year'],data['own'])

#add title to plot
plt.title("scatter plot")

#setting x and y axis 
plt.xlabel('year')
plt.ylabel('own')

#result
plt.show()
```

```python
#line chart
plt.plot(data['year'])
plt.plot(data['own'])

plt.title("line chart")

plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
#bar chart 
plt.bar(data['year'],data['own'])

plt.title("bar chart")

plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
#histogram
plt.hist(data['income'])

plt.title("histogram")

plt.show()
```
